### Data Types
Python data types can be grouped into Primitive and Non-Primitive types:
#### Primitive Data Types

These store a single value.

**int** – Integer (e.g., 10)

**float** – Decimal number (e.g., 3.14)

**complex** – Complex number (e.g., 2+3j)


**bool** – Boolean (True, False)

**str** – String/Text (e.g., "Hello")

#### Non-Primitive Data Types

These can store multiple values.


**list** – Ordered, mutable collection (e.g., [1, 2, 3])

**tuple** – Ordered, immutable collection (e.g., (1, 2, 3))

**set** – Unordered collection of unique items (e.g., {1, 2, 3})

**dict** – Key-value pairs (e.g., {"name": "Ali", "age": 20})

## Conditional Statements
Conditional statements allow your program to make decisions.
#### IF Statement
Used when you want to check one condition.

```python
age = 20

if age >= 18:
    print("You can vote")
```

#### IF ELSE
Used when there are two possibilities.
#### IF ELIF ELSE
Used when there are multiple conditions.

```python
marks = 85

if marks >= 90:
    print("Grade A+")

elif marks >= 80:
    print("Grade A")

elif marks >= 70:
    print("Grade B")

else:
    print("Fail")
```

Python checks conditions from top to bottom. Once one is true, it stops checking the rest.

#### Nested IF ELSE
Nested means an if statement inside another if statement.

```python
age = 22
has_license = True

if age >= 18:

    if has_license:
        print("You can drive")

    else:
        print("Get a driving license")

else:
    print("You are too young")
```

## Student Information System
This program is a Student Information System. It takes information from the user, 
checks the student's age and marks, then displays the appropriate grade.

```python
# Student Information System

name = input("Enter your name: ")
age = int(input("Enter your age: "))
marks = float(input("Enter your marks: "))

if age >= 18:

    if marks >= 90:
        grade = "A+"

    elif marks >= 80:
        grade = "A"

    elif marks >= 70:
        grade = "B"

    elif marks >= 60:
        grade = "C"

    else:
        grade = "Fail"

    print("\nStudent Report")
    print("Name:", name)
    print("Age:", age)
    print("Marks:", marks)
    print("Grade:", grade)

else:
    print("Student is under 18.")
```
